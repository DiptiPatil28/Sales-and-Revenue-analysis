```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Create Sample Sales Dataset
data = {
    "Date": pd.date_range(start="2024-01-01", periods=10, freq="ME"),
    "Product": ["Laptop", "Mobile", "Tablet", "Laptop", "Mobile",
                "Tablet", "Laptop", "Mobile", "Tablet", "Laptop"],
    "Region": ["North", "South", "East", "West", "North",
               "South", "East", "West", "North", "South"],
    "Revenue": [50000, 30000, 20000, 55000, 32000,
                21000, 60000, 35000, 23000, 65000]
}

df = pd.DataFrame(data)
df
```

```python
#Basic Data Check
df.info()
df.describe()
```

```python
#Total Revenue
total_revenue = df["Revenue"].sum()
print("Total Revenue:", total_revenue)
```

```python
#Product-wise Revenue Analysis
product_sales = df.groupby("Product")["Revenue"].sum()
product_sales
```

```python
#Bar Chart
product_sales.plot(kind="bar")
plt.title("Product-wise Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.show()
```

```python
#Region-wise Revenue Analysis
region_sales = df.groupby("Region")["Revenue"].sum()
region_sales
```

```python
#Pie Chart
region_sales.plot(kind="pie", autopct="%1.1f%%")
plt.title("Region-wise Revenue Distribution")
plt.ylabel("")
plt.show()
```

```python
#Monthly Revenue Trend
monthly_sales = df.groupby(df["Date"].dt.month)["Revenue"].sum()
monthly_sales
```

```python
#Line chart
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

**KEY INSIGHTS**

 Laptops generate the highest revenue among all products.

 North and South regions contribute the most to total sales.

 Revenue shows an overall increasing trend over time.
